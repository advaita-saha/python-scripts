# Find graph of a function and it's derivative


### Below example for function 
$ f(x) = x^2 + x + 1 $

```python
# Program will workout the plot if you change return statement for you required function

def f(x):
    return x**2 + x + 1
def df(x):
    return 2*x +1
```

```python
#imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_1 = np.linspace(start=-3.5,stop = 2.5,num = 100)
```

```python
# to show x,y axis
axis = np.linspace(-20,20,5)
```

## Plot of Function and It's Derivative

```python
plt.figure(figsize=[15,6]) 
#subplot1
plt.subplot(1,2,1)
plt.plot(axis*0,axis,color="black")   #y-axis
plt.plot(axis,axis*0,color="black")   #x-axis
plt.plot(x_1,f(x_1))
plt.xlabel("X", fontsize=16)
plt.ylabel("F(x)",fontsize =16)
plt.title("f(x)=x^2+x+1")
plt.xlim(-3.9,3.5)
plt.ylim(-12,24)
plt.grid()

#subplot 2
plt.subplot(1,2,2)
plt.plot(axis*0,axis,color="black")   #y-axis
plt.plot(axis,axis*0,color="black")   #x-axis
plt.plot(x_1,df(x_1))
plt.xlim(-3.9,3.5)
plt.ylim(-12,12)
plt.grid()
plt.show()
```
